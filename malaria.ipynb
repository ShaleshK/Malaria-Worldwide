```python
df = pd.read_csv('data/reported_indigenous_confirmed_cases.csv')
popn_df = pd.read_csv('data/population_stats.csv')
```

```python
df_clean = df[['year', 'name', 'value', 'REGION (CODE)']]
df_clean.head().T
```

```python
same = lambda x:x
```

```python
df_grouped = df.groupby(['name','year']).agg({'name':same,'year':same,"value": "sum"})
```

```python
df_grouped['lastValue'] = df_grouped['value'].shift(periods=1)
```

```python
df_grouped.reset_index(drop=True,inplace=True)
```

```python
df_grouped.head()
```

```python
df_grouped.sort_values(["name","year"],ascending=False).head()
```

```python
df_grouped['rank'] = df.groupby(["year" ])["value"].rank(ascending = True, method = 'first')
df_grouped.loc[df_grouped['year'] == 2000].head()
```

```python
df_grouped.sort_values(["name","year"],ascending=True).head()
```

```python
mydict = {}
for index, row in df_grouped.iterrows():
    name = row['name']
    value = row['value']
    if value == "NAN":
        row['value'] =   mydict[name]
    else:
        mydict[name] = value
    
            
        
```

```python
# df_grouped["value"].value_counts()
# for n in df_grouped['name'].unique():
#     for y in df_grouped['year'].unique:
# find all countries
countries = df_grouped['name'].unique()
# instantiate final df
df_final = pd.DataFrame( columns=["name","year","value","lastValue","rank"])
# dataframe will go from 2000-2014
# iterate through countries

for country in countries:
    # instantiate datafrome for each country
    df_country = pd.DataFrame()
    # filter original dataframe for this country
    df_filter = df_grouped.loc[df_grouped['name'] == country]
    # make year our index in filtered dataframe
    df_filter = df_filter.set_index(['year'])
# instantiate store for latest value that we see for this country
    cases = df_filter.value.mean()
    rank = 0
# iterate through 2000-2014
    for year in range(2000, 2015):
        if not year in df_filter.index:
            df_temp = pd.DataFrame([[country,year,cases,cases,rank]], columns=["name","year","value","lastValue","rank"])
            df_final= df_final.append(df_temp, ignore_index=True)

            
```
